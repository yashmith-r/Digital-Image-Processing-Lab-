```python
#Perform the following operations using library functions

#a. Read, Display and write any color image in other formats.
#b. Find RED, GREEN and BLUE plane of the color image.
#c. Convert color image to grayscale image and binary image
#d. Resize the image by one half and one quarter.
#e. Rotates Image  by 45, 90 and 180 degrees and display
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#a. Read, Display and write any color image in other formats.
img = Image.open('Naruto_2.jpg')
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()

img.save('output_image.png')
img.save('output_image.bmp')
img.save('output_image.tiff')

print("Image saved in different formats: PNG, BMP, and TIFF")
```

```python
#b. Find RED, GREEN and BLUE plane of the color image.

I = np.array(Image.open('Naruto_2.jpg'))

r, c, _ = I.shape

R = np.zeros((r, c, 3), dtype=np.uint8)
G = np.zeros((r, c, 3), dtype=np.uint8)
B = np.zeros((r, c, 3), dtype=np.uint8)


R[:, :, 0] = I[:, :, 0]
G[:, :, 1] = I[:, :, 1]
B[:, :, 2] = I[:, :, 2]


plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(R)
plt.title('Red Plane')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(G)
plt.title('Green Plane')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(B)
plt.title('Blue Plane')
plt.axis('off')

plt.show()
```

```python
gray_img = img.convert('L')

threshold = 127
binary_img = gray_img.point(lambda x: 255 if x > threshold else 0, mode='1')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

plt.show()
```

```python
#d. Resize the image by one half and one quarter.
half_img = img.resize((img.width // 2, img.height // 2))
quarter_img = img.resize((img.width // 4, img.height // 4))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(half_img)
plt.title('Half Size Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(quarter_img)
plt.title('Quarter Size Image')
plt.axis('off')
plt.show()
```

```python
r45 = img.rotate(45, expand=True)
r90 = img.rotate(90, expand=True)
r180 = img.rotate(180, expand=True)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(r45)
plt.title('Rotated 45 Degrees')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(r90)
plt.title('Rotated 90 Degrees')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(r180)
plt.title('Rotated 180 Degrees')
plt.axis('off')

plt.show()
```
